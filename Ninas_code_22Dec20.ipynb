```python
import csv
import pandas as pd
import numpy
import scipy.stats as st
from scipy.stats import linregress
import time
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('ultimate.csv')
```

```python
data.head()
```

```python
clean = data.drop(columns = ["Unnamed: 0"])
clean.head()
clean['Infection_rate'] = clean['Infection_rate'].str.rstrip('%').astype('float')
new_df = clean.dropna()
```

```python
x = new_df["stringency_index"]
y = new_df["Infection_rate"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x, y)
regress_values = x * slope + intercept
#calculate regression line using standard equation
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(6,10),fontsize=15,color="red")
plt.scatter(x,y)
plt.show(line_eq, rvalue)
print(line_eq)
```
